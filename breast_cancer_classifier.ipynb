importing libraries and dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
raw_data= pd.read_csv('/content/data.csv')
```

```python
raw_data.head(10)
```

```python
raw_data=raw_data.drop(columns=['id','Unnamed: 32'],axis=1)
```

```python
raw_data.info()
```

```python
raw_data.isnull().sum()
```

```python
raw_data.describe()
```

```python
raw_data.shape
```

```python
raw_data['diagnosis'].replace({'M':1 , 'B':0},inplace=True)
```

```python
raw_data.head()
```

```python
raw_data.isnull().sum()
```

```python
X = raw_data.drop(columns=['diagnosis'])
Y = raw_data['diagnosis']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2,random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train_normalized = scaler.transform(X_train)
print(X_train_normalized)
```

```python
X_test_normalized= scaler.transform(X_test)
print(X_test_normalized)
```

```python
model= LogisticRegression()
model.fit(X_train_normalized,Y_train)
```

```python
#evalaution of model

prediction_on_training_data = model.predict(X_train_normalized)
accuracy_train = accuracy_score(Y_train,prediction_on_training_data)
print(accuracy_train)
```

```python
prediction_on_testing_data = model.predict(X_test_normalized)
accuracy_test = accuracy_score(Y_test,prediction_on_testing_data)
print(accuracy_test)
```
